# Sprzedaż w poszczególnych kanałach i dniach tygodnia

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
```

```python
import cufflinks as cf
cf.go_offline()
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
import pandas as pd
import datetime as dt

t16 = pd.read_csv('transactions-2016.csv',parse_dates=True)
t17 = pd.read_csv('transactions-2017.csv',parse_dates=True)
t18 = pd.read_csv('transactions-2018.csv',parse_dates=True)
t19 = pd.read_csv('transactions-2019.csv',parse_dates=True)
```

```python
tran = pd.concat([t16,t17,t18,t19])
tran.head()
```

```python
tran['tran_date'] = pd.to_datetime(tran['tran_date'])        
week_tran = tran.groupby(['Store_type',tran['tran_date'].dt.day_name()]).sum()
week = week_tran[['Qty']]
week1 = week.reset_index()
week1.head()
```

```python
week1.iplot(kind='scatter',
x='tran_date',
y='Qty',
mode='markers',
xTitle='weekday',
yTitle='Qty',
title='Wielkość sprzedaży w dniach tygodnia i kanałach sprzedaży',
text='Store_type')
```

Odp. Największa sprzedaż pod względem ilościowym ma miejsce w weekendy w e-Shopie.

```python
tran['tran_date'] = pd.to_datetime(tran['tran_date'])        
week_tran = tran.groupby(['Store_type',tran['tran_date'].dt.day_name()]).sum()
week1 = week_tran[['Rate']]
week2 = week1.reset_index()
```

```python
week2.iplot(kind='scatter',
x='tran_date',
y='Rate',
mode='markers',
xTitle='weekday',
yTitle='rate',
title='Wartość sprzedaży w dniach tygodnia i kanałach sprzedaży',
text='Store_type')
```

Odp. Największa sprzedaż pod względem wartościowym ma miejsce w soboty w e-Shopie.

```python
week2.set_index('Store_type')['Rate'].sort_values(ascending=False).iplot(kind='bar')
```

Odp. Największa sprzedaż pod względem wartościowym ma miejsce w e-Shopie. W każdym z pozostałych trzech kanałów sprzedaż jest o połowę mniejsza.

# Informacje o klientach każdego z kanałów sprzedaży

```python
import pandas as pd
import datetime as dt

pro = pd.read_csv('prod_cat_info.csv')
cus = pd.read_csv('customers.csv',parse_dates=True)
cou = pd.read_csv('country_codes.csv')
```

```python
pro.head()
```

```python
cus.head()
```

```python
cou1 = cou.rename(columns={'Unnamed: 0':'country_code'})
cou1.head()
```

```python
mer1 = pd.merge(cus,cou1,how='inner',on='country_code')
mer2 = mer1.rename(columns={'customer_Id':'cust_id'})
```

```python
tran1 = pd.merge(tran,mer2,how='inner',on='cust_id')
```

```python
tran2 = pd.merge(tran1,pro,how='inner',on='prod_cat_code')
tran2.head()
```

```python
from plotly import subplots
```

```python
tran1.set_index('Gender')['Rate'].sort_values(ascending=False).iplot(kind='bar')
```

Odp. Płeć nie ma większego znaczenia dla wartości sprzedaży, nieznacznie więcej wydali mężczyźni.

```python
tran3 = tran2.groupby(pd.Grouper(key='prod_cat')).agg({'Rate':'sum'})
tran4 = tran3.reset_index()
tran4.head()
```

```python
import matplotlib.pyplot as plt

labels = tran4['prod_cat']
sizes = tran4['Rate']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

Odp. Klienci najwięcej wydali na produkty w kategorii Books - prawie 37% wartości sprzedaży sklepu, Electronics - ponad 25% i Home and kitchen - ponad 16%. Wartość sprzedaży produktów z każdej z kategorii Footware, Clothing i Bags nie przekroczyła 10% wartości sprzedaży sklepu.

```python
tran5 = tran2.groupby(pd.Grouper(key='prod_subcat')).agg({'Rate':'sum'})
tran6 = tran5.reset_index()
tran6.head()
```

```python
tran6.set_index('prod_subcat')['Rate'].sort_values(ascending=False).iplot(kind='bar')
```

Odp. Klienci najwięcej wydają na produkty z podkategorii Mens i Women, natomiast najmniej na Furnishing, Bath, Tools i Kitchen.

```python
fig = go.Figure()

fig.add_trace(go.Choropleth(colorscale='Viridis',
                           reversescale=True,
                           locations=tran1['country'],
                           locationmode='country names',
                           z = tran1['Rate'],
                           colorbar = dict(title='rate')))

fig.update_layout(title='Wartość sprzedaży w podziale na kraje',
                 margin={'r':30,'t':50,'l':0,'b':30},
                 geo=dict(showframe=False,projection={'type':'mercator'}))
iplot(fig)
```

```python
tran7 = tran1.groupby(pd.Grouper(key='country')).agg({'Rate':'sum'})
tran8 = tran7.reset_index()
tran8.sort_values(by='Rate',ascending=False)
```

Odp. Wartość sprzedaży w poszczególnych krajach jest zbliżona, natomiast największą sprzedaż wygenerowali klienci z Francji, a najmniejszą z Norwegii.
